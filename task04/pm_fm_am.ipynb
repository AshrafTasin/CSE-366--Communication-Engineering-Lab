#Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

#Phase Modulation (PM) :

```python
t=np.arange(0,2,0.001)
fc=10
fm=1
phase_deviation=np.pi/2

mt=0.5*(1+signal.square(2*np.pi*fm*t))
ct=np.sin(2*np.pi*fc*t)
pm=np.sin(2*np.pi*fc*t+phase_deviation*mt)


plt.subplot(3,1,1)
plt.title('Phase Modulation')
plt.plot(t,ct,'g')
plt.ylabel('Amplitude')
plt.xlabel('Carrier signal')
plt.grid('on')

plt.subplot(3,1,2)
plt.plot(t,mt, 'r')
plt.ylabel('Amplitude')
plt.xlabel('Message signal')
plt.grid('on')

plt.subplot(3,1,3)
plt.plot(t,pm, color="purple")
plt.ylabel('Amplitude')
plt.xlabel('PM signal')
plt.grid('on')

plt.subplots_adjust(hspace=1)
plt.rc('font', size=10)
fig = plt.gcf()
fig.set_size_inches(16, 9)

fig.savefig('Phase Modulation.png', dpi=100)
```

#Frequency Modulation (FM) :

```python
modulator_frequency = 5.0
carrier_frequency = 50.0
modulation_index = 2.0

time = np.arange(44100.0) / 44100.0
modulator = np.sin(2.0 * np.pi * modulator_frequency * time) * modulation_index
carrier = np.sin(2.0 * np.pi * carrier_frequency * time)
product = np.zeros_like(modulator)

for i, t in enumerate(time):
    product[i] = np.sin(2. * np.pi * (carrier_frequency * t + modulator[i]))


plt.subplot(3,1,1)
plt.title('Frequency Modulation')
plt.plot(modulator,'g')
plt.ylabel('Amplitude')
plt.xlabel('Modulator')
plt.grid('on')

plt.subplot(3,1,2)
plt.plot(carrier, 'r')
plt.xlabel('Carrier')
plt.ylabel('Amplitude')
plt.grid('on')

plt.subplot(3,1,3)
plt.plot(product, color="purple")
plt.ylabel('Amplitude')
plt.xlabel('FM signal')
plt.grid('on')

plt.subplots_adjust(hspace=1)
plt.rc('font', size=10)
fig = plt.gcf()
fig.set_size_inches(16, 9)

fig.savefig('Frequency Modulation.png', dpi=100)
```

#Amplitude Modulation (AM) :

```python
A_c=10
f_c=50
A_m=5
f_m=30
t = np.linspace(0, 1, 1000)

carrier = A_c*np.cos(2*np.pi*f_c*t)
modulator = A_m*np.cos(2*np.pi*f_m*t)
product = A_c*(1+modulation_index*np.cos(2*np.pi*f_m*t))*np.cos(2*np.pi*f_c*t)

plt.subplot(3,1,1)
plt.title('Amplitude Modulation')
plt.plot(modulator,'g')
plt.ylabel('Amplitude')
plt.xlabel('Message signal')

plt.subplot(3,1,2)
plt.plot(carrier, 'r')
plt.ylabel('Amplitude')
plt.xlabel('Carrier signal')

plt.subplot(3,1,3)
plt.plot(product, color="purple")
plt.ylabel('Amplitude')
plt.xlabel('AM signal')

plt.subplots_adjust(hspace=1)
plt.rc('font', size=15)
fig = plt.gcf()
fig.set_size_inches(16, 9)

fig.savefig('Amplitude Modulation.png', dpi=100)
```
