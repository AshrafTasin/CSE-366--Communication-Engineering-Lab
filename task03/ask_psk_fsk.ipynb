#Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
```

#ASK

```python
f1=60
f2=5
a=5

t=np.arange(0,1,1/1000)
x=a*np.sin(2*np.pi*f1*t)+(a/2)
u=(a/2)*signal.square(2 * np.pi * f2 * t)+(a/2)
v=x*u


plt.subplot(3,1,1)
plt.title('Amplitude Shift Keying')
plt.plot(x,'g')
plt.ylabel('Amplitude')
plt.xlabel('Carrier')
plt.grid('on')

plt.subplot(3,1,2)
plt.plot(t,u, 'r')
plt.ylabel('Amplitude')
plt.xlabel('Message signal')
plt.grid('on')

plt.subplot(3,1,3)
plt.plot(t,v, color="purple")
plt.ylabel('Amplitude')
plt.xlabel('ASK signal')
plt.grid('on')

plt.subplots_adjust(hspace=1)
plt.rc('font', size=10)
fig = plt.gcf()
fig.set_size_inches(16, 9)

fig.savefig('Amplitude Shift Keying.png', dpi=100)
```

#PSK

```python
f1=30
f2=5
a=5

t=np.arange(0,1,1/1000)
x=a*np.sin(2*np.pi*f1*t)
u=signal.square(2 * np.pi * f2 * t)
v=x*u


plt.subplot(3,1,1)
plt.title('Phase Shift Keying')
plt.plot(t,x,'g')
plt.ylabel('Amplitude')
plt.xlabel('Carrier Signal')
plt.grid('on')

plt.subplot(3,1,2)
plt.plot(t,u, 'r')
plt.ylabel('Amplitude')
plt.xlabel('Message signal')
plt.grid('on')

plt.subplot(3,1,3)
plt.plot(t,v, color="purple")
plt.ylabel('Amplitude')
plt.xlabel('PSK signal')
plt.grid('on')

plt.subplots_adjust(hspace=1)
plt.rc('font', size=10)
fig = plt.gcf()
fig.set_size_inches(16, 9)

fig.savefig('Phase Shift Keying.png', dpi=100)
```

#FSK

```python
f1=50
f2=0.4
amplitude=10
f=5

t=np.arange(0,1,1/1000)
x=amplitude*np.sin(2*np.pi*f1*t)+1
u=signal.square(2*np.pi*f*t)
fsk=f1+f1*u/f2
x2=amplitude*np.sin(2*np.pi*fsk*t)


plt.subplot(3,1,1)
plt.title('Frequency Shift Keying')
plt.plot(t,x,'g')
plt.ylabel('Amplitude')
plt.xlabel('Carrier Signal')
plt.grid('on')

plt.subplot(3,1,2)
plt.plot(u, 'r')
plt.ylabel('Amplitude')
plt.xlabel('Message signal')
plt.grid('on')

plt.subplot(3,1,3)
plt.plot(t,x2, color="purple")
plt.ylabel('Amplitude')
plt.xlabel('FSK signal')
plt.grid('on')

plt.subplots_adjust(hspace=1)
plt.rc('font', size=10)
fig = plt.gcf()
fig.set_size_inches(16, 9)

fig.savefig('Frequency Shift Keying.png', dpi=100)
```
